This is a simple simulation to consider 3 cases in which we cancel the GPs and check how much it would have affected the championship

```python
import pandas as pd
from matplotlib.pyplot import plot as plt
```

```python
%matplotlib inline
```

Here is a simple function to allow us to visualize points instead of the ranking

```python
def convert_ranking_to_points(rank):
    """
    Original dataset provides the rank instead of points,
    but we are interested in the points only
    """
    points = [25, 18, 15, 12, 10, 8, 6, 4, 2, 1]
    try:
        rank_index = int(rank)-1
    except ValueError:
        # "Ret" = retired the car/no points
        rank_index = 10
    try:
        return points[rank_index]
    except IndexError:
        # no points
        return 0
```

## Real 2019 standings

```python
# data from https://www.f1-fansite.com/f1-results/2019-f1-championship-standings/
df = pd.read_csv('f1_2019.csv')
gps = df.columns[2:-1] # holds gp info 
for gp in gps:
    df[gp] = df[gp].apply(convert_ranking_to_points)
df["TOTAL"] = df[gps].sum(axis=1) # we are not considering the fastest lap
```

### 2019 RESULTS

```python
df.sort_values(by=['PTS']).plot.barh(x="DRIVER", y="PTS", figsize=(9,8))
```

## CASE 1
### CANCELLED: AUSTRALIA

```python
df_case1 = df
df_case1["TOTAL"] = df[gps.drop("AUS")].sum(axis=1) # we are not considering the fastest lap
df_case1.sort_values(by=['TOTAL']).plot.barh(x="DRIVER", y="TOTAL", figsize=(9,8))
```

```python
df_case1.sort_values(by=['TOTAL'], ascending=False)
```

## CASE 2
### CANCELLED: AUSTRALIA, BAHRAIN, CHINA

Now, let's consider if the postponed GPs also do get cancelled

```python
df_case2 = df
df_case2["TOTAL"] = df[gps.drop(["AUS", "BAH", "CHI"])].sum(axis=1) # we are not considering the fastest lap
df_case2.sort_values(by=['TOTAL']).plot.barh(x="DRIVER", y="TOTAL", figsize=(9,8))
```

```python
df_case2.sort_values(by=['TOTAL'], ascending=False)
```

## CASE 3
### CANCELLED: AUSTRALIA, BAHRAIN, CHINA, AZERBAIJAN, SPAIN, MONACO

In this case, we are considering the same time-gap as we expect for the outbreak to die off, taking into consideration the 2019 calendar

```python
df_case3 = df
df_case3["TOTAL"] = df[gps.drop(["AUS", "BAH", "CHI", "AZE", "SPA", "MON"])].sum(axis=1) # we are not considering the fastest lap
df_case3.sort_values(by=['TOTAL']).plot.barh(x="DRIVER", y="TOTAL", figsize=(9,8))
```
